#### Data Processing

Check Null Values and Duplicates: No null values or duplicates were found. No need to drop any rows.

```python
print(df.isnull().sum())
```

```python
# Show duplicate rows
duplicated_rows = df[df.duplicated()]
print("Duplicated Rows:")
print(duplicated_rows)
```

```python
# Show duplicate values in 'id' column
duplicate_ids = df[df.duplicated('id', keep=False)]
print("Duplicate IDs:")
print(duplicate_ids)
```

Ensure correct data types for each column

```python
# Convert from string to list and remove square brackets and quotation signs
df['artists'] = df['artists'].apply(lambda x: ', '.join(ast.literal_eval(x)) if isinstance(x, str) else x) 
df['duration_ms'] = df['duration_ms'].astype(int)
df['explicit'] = df['explicit'].astype(int)
df['id'] = df['id'].astype(str)
df['key'] = df['key'].astype(int)
df['mode'] = df['mode'].astype(int)
df['name'] = df['name'].astype(str)
df['popularity'] = df['popularity'].astype(int)
df['release_date'] = df['release_date'].astype(str) 
df['year'] = df['year'].astype(int)
```

Convert 'duration_ms' from milliseconds to seconds

```python
df['duration_sec'] = df['duration_ms'] / 1000
```

Round numerical columns to 3 decimal point unless it's extremely small value

```python
# Round numerical columns
numeric_cols = ['acousticness', 'danceability', 'energy', 'instrumentalness', 
                'liveness', 'loudness', 'speechiness', 'tempo', 'valence']

# Apply function to format and round up
def format_and_round(x):
    # Ensure the value is not NaN
    if pd.notnull(x):  
        # Convert to float with 10 decimal places to avoid scientific notation
        x = float(f"{x:.10f}")
        # If the absolute value is >= 0.001, round it to 3 decimal places
        if abs(x) >= 0.001:
            return round(x, 3)
        else:
            return x  # Leave small values as they are
    return x  # Return NaN or non-numeric values unchanged

df[numeric_cols] = df[numeric_cols].apply(lambda col: col.apply(format_and_round))
```

Extract month from 'release_date' column

```python
df['month'] = df['release_date'].apply(lambda x: x[5:7] if '-' in x else '')
```
